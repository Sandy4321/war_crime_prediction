```python
# df.dropna(subset=['Date'], inplace=True)
df['Epoch'] = df.Date.astype(int) // 10**9
```

```python
gen_loc = df[['Gender', 'Date', 'MonthDate', 'WeekDate', 'Epoch', 'Admin3EN', 'GovernateEN', 'Location', 'Lon', 'Lat']]
```

```python
gen_loc = pd.concat([pd.get_dummies(gen_loc.Gender), gen_loc], axis=1)
```

```python
g = gen_loc.groupby(['MonthDate', 'Admin3EN'])
prop = g.sum()[['Female','Male']]
prop['Female_proportion'] = prop.apply(lambda x: x.Female / x.sum(), axis=1)
# prop.reset_index(inplace=True)
```
